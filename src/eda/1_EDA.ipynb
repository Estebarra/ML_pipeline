```python
# Crear variables con el conteo de cada valor de cada columna

performance_count = student_entry_performance_df['Performance'].value_counts()
gender_count = student_entry_performance_df['Gender'].value_counts()
caste_count = student_entry_performance_df['Caste'].value_counts()
coaching_count = student_entry_performance_df['coaching'].value_counts()
time_count = student_entry_performance_df['time'].value_counts()
class_ten_education_count = student_entry_performance_df['Class_ten_education'].value_counts()
twelve_education_count = student_entry_performance_df['twelve_education'].value_counts()
medium_count = student_entry_performance_df['medium'].value_counts()
class_X_percentage_count = student_entry_performance_df['Class_ X_Percentage'].value_counts()
class_XII_percentage_count = student_entry_performance_df['Class_XII_Percentage'].value_counts()	
father_occupation_count = student_entry_performance_df['Father_occupation'].value_counts()
mother_occupation_count = student_entry_performance_df['Mother_occupation'].value_counts()
```

```python
# Visualización de los datos

plt.figure(figsize=(14, 60))

plt.subplot(6, 2, 1)
performance_count.plot(kind='bar', color='firebrick')
plt.title('Performance')
plt.ylabel('Count')

plt.subplot(6, 2, 2)
gender_count.plot(kind='bar', color='gold')
plt.title('Gender')
plt.ylabel('Count')

plt.subplot(6, 2, 3)
caste_count.plot(kind='bar', color='chartreuse')
plt.title('Caste')
plt.ylabel('Count')

plt.subplot(6, 2, 4)
coaching_count.plot(kind='bar', color='turquoise')
plt.title('coaching')
plt.ylabel('Count')

plt.subplot(6, 2, 5)
time_count.plot(kind='bar', color='navy')
plt.title('time')
plt.ylabel('Count')

plt.subplot(6, 2, 6)
class_ten_education_count.plot(kind='bar', color='magenta')
plt.title('Class_ten_education')
plt.ylabel('Count')

plt.subplot(6, 2, 7)
twelve_education_count.plot(kind='bar', color='salmon')
plt.title('twelve_education')
plt.ylabel('Count')

plt.subplot(6, 2, 8)
medium_count.plot(kind='bar', color='wheat')
plt.title('medium')
plt.ylabel('Count')

plt.subplot(6, 2, 9)
class_X_percentage_count.plot(kind='bar', color='palegreen')
plt.title('Class_ X_Percentage')
plt.ylabel('Count')

plt.subplot(6, 2, 10)
class_XII_percentage_count.plot(kind='bar', color='aqua')
plt.title('Class_XII_Percentage')
plt.ylabel('Count')

plt.subplot(6, 2, 11)
father_occupation_count.plot(kind='bar', color='blueviolet')
plt.title('Father_occupation')
plt.ylabel('Count')

plt.subplot(6, 2, 12)
mother_occupation_count.plot(kind='bar', color='crimson')
plt.title('Mother_occupation')
plt.ylabel('Count')
```
